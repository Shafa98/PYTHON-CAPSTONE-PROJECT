# Data Cleaning


```python
#import library
import pandas as pd
import numpy as np
import re
from datetime import datetime
```

```python
#import data and choose usage sheet
data=pd.read_excel("C:\\Users\\99451\\Downloads\\data_cleaning.xlsx", sheet_name="Usage")
```

```python
#show data
data.head(10)
```

```python
#give column name to data
column_names=['day']
```

```python
data.columns=column_names
```

```python
#show data again
data
```

```python
# put comma after am and pm
data["day"] = data["day"].str.replace("AM", "AM,")
data["day"] = data["day"].str.replace("PM", "PM,")
```

```python
# show the first 10 rows of the DataFrame
data.head(10)
```

```python
#split day column in two column using comma
data = pd.DataFrame(data.day.str.split(',',1).tolist(),
                                 columns = ['time','row'])
```

```python
#see result
data
```

```python
#replace underscore with space for removing underscore
data['time']=data['time'].str.replace("_", "")
```

```python
# do the same for row column
data['row']=data['row'].str.replace("_+", ",")
```

```python
#see data
data
```

```python
#split row column in two new columns for using comma
data[['date', 'elec_usage']] = data['row'].str.split(',', n=1, expand=True)
```

```python
#delete row column
data.drop(columns=['row'], inplace=True)
```

```python
# display data
data
```

```python
# in some rows there is no space between number and am/pm #put space with regex
def add_spaces_to_time(input_string):
    return re.sub(r'(\d+)([APap][Mm])', r'\1 \2', input_string)
```

```python
#apply function to time column
data['time'] = data['time'].apply(add_spaces_to_time)
```

```python
#see result
data
```

```python
#in date column there is th, nd, rd, st #remove them with regex
data['date'] = data['date'].str.replace(r'\b(\d+)(?:st|nd|rd|th)\b', r'\1', regex=True).str.strip()
```

```python
#see result
data
```

```python
#find weekdays, use regex
pattern = r'\b(Mon(?:day)?|Tue(?:sday)?|Wed(?:nesday)?|Thu(?:rsday)?|Fri(?:day)?|Sat(?:urday)?|Sun(?:day)?)'
```

```python
#create weekdays column, extract weekdays from date column
data['weekdays'] = data['date'].str.extract(pattern)
```

```python
# Remove weekdays from the 'date' column using regex
data['date'] = data['date'].str.replace(pattern, '').str.strip()
```

```python
#see result
data
```

```python
#in some column, weekdays have given like full name, but some columns as short version. Do it short version with using regex
pattern = r'\b(Mon|Tue|Wed|Thu|Fri|Sat|Sun)\b'
    
```

```python
#apply it weekdays column and extract pattern
data['weekdays'] = data['weekdays'].str.extract(pattern, flags=re.IGNORECASE)
```

```python
#see result
data
```

```python
#change date column type, convert it date format
data['date'] = pd.to_datetime(data['date'])
```

```python
#for filling NAN's from weekdays column use day from date column
data['weekdays'] = data['weekdays'].fillna(data['date'].dt.strftime('%a'))
```

```python
#see result
data
```

```python
#change time column format
def convert_to_24_hour(time_str):
    time_str = ''.join(time_str.split())
    time_obj = datetime.strptime(time_str, '%I%p')
    return time_obj.strftime('%H:%M')
```

```python
#apply it to time column
data['time'] = data['time'].apply(convert_to_24_hour)
```

```python
#see result
data
```

```python
#extract kwh from elec usage column and create new columns
data[['elec_usage', 'measure']] = data['elec_usage'].str.extract(r'(\d+\.\d+)\s*(\w*)')
```

# Question 5 Based on your historic electricity usage, what would your annual cost of electricitybe under the "Monthly Flex" contract?
# Function to calculate the cost for each row

```python

def calculate_cost(row):
    month = pd.to_datetime(row['date']).month
    cost_per_kwh = monthly_flex_rates[pd.to_datetime(row['date']).strftime('%B')]
    return row['elec_usage'] * cost_per_kwh
```

```python
# Calculate the cost for each row
electricity_data1= data.apply(calculate_cost, axis=1)
```

```python
# Calculate the total annual cost
annual_cost = electricity_data1.sum()
print("Annual cost of electricity under Monthly Flex contract: ${:.2f}".format(annual_cost)) #answer is b. $1420.06
```

# Question 6: Based on your historic electricity usage, which of the three contracts would pro- duce the lowest annual cost?

```python
# Modify flat rate 
flat_rate = 0.21
```

```python
# Modify monthly_flex_rates dictionary
monthly_flex_rates = {
    'January': 0.20,
    'February': 0.19,
    'March': 0.17,
    'April': 0.18,
    'May': 0.22,
    'June': 0.27,
    'July': 0.24,
    'August': 0.19,
    'September': 0.18,
    'October': 0.15,
    'November': 0.14,
    'December': 0.19,
}
```

```python
# modify hourly_flex_rates dictionary
hourly_flex_rates = {
    '00:00': 0.10,
    '01:00': 0.10,
    '02:00': 0.10,
    '03:00': 0.10,
    '04:00': 0.15,
    '05:00': 0.20,
    '06:00': 0.24,
    '07:00': 0.24,
    '08:00': 0.26,
    '09:00': 0.20,
    '10:00': 0.20,
    '11:00': 0.26,
    '12:00': 0.26,
    '13:00': 0.20,
    '14:00': 0.24,
    '15:00': 0.18,
    '16:00': 0.15,
    '17:00': 0.30,
    '18:00': 0.24,
    '19:00': 0.24,
    '20:00': 0.12,
    '21:00': 0.11,
    '22:00': 0.10,
    '23:00': 0.10,
}
```

```python
# total annual cost for each contract
def calculate_annual_cost(rate):
    return sum(rate * usage for usage in data['elec_usage']) * 365
```

```python
flat_rate_cost = calculate_annual_cost(flat_rate)
```

```python
# Convert 'date' to month format 
months = [date.strftime('%B') for date in data['date']]
```

```python
# total annual cost for Monthly Flex contract
monthly_flex_cost = sum(monthly_flex_rates[month] * usage for month, usage in zip(months, data['elec_usage']))
monthly_flex_cost *= 365
```

```python
hourly_flex_cost = sum(hourly_flex_rates[time] * usage for time, usage in zip(data['time'], data['elec_usage']))
hourly_flex_cost *= 365
```

```python
# Calculate total annual cost for Hourly Flex contract
hourly_flex_cost = sum(hourly_flex_rates[time] * usage for time, usage in zip(data['time'], data['elec_usage']))
hourly_flex_cost *= 365
```

```python
# the contract with the lowest annual cost
lowest_cost_contract = min(flat_rate_cost, monthly_flex_cost, hourly_flex_cost)
```

```python
# write the contract name
contract_names = {
    flat_rate_cost: 'Flat Rate',
    monthly_flex_cost: 'Monthly Flex',
    hourly_flex_cost: 'Hourly Flex',
}
```

```python
lowest_cost_contract_name = contract_names[lowest_cost_contract]
```

```python
# Print result
print(f"The contract that would produce the lowest annual cost is: {lowest_cost_contract_name}") #answer is c
```
